```python
import requests, pandas as pd, datetime as dt

def fetch_preclose(date: dt.date) -> pd.DataFrame:
    d = date.strftime("%Y%m%d")
    url = f"https://www.twse.com.tw/exchangeReport/STOCK_DAY_ALL"
    payload = {"response": "json", "date": d} # Date not working
    js = requests.get(url, params=payload, timeout=15).json()
    df = pd.DataFrame(js["data"], columns=js["fields"])
    df["收盤價"] = pd.to_numeric(df["收盤價"], errors="coerce")
    df["漲跌價差"] = (df["漲跌價差"]
                      .str.replace("X", "", regex=False)
                      .str.replace(",", "", regex=False)
                      .astype(float))
    df["pre_close"] = df["收盤價"] - df["漲跌價差"]
    return df[["證券代號", "pre_close"]]

# 例：抓 2025-07-14 的前收
pc = fetch_preclose(dt.date(2025, 7, 14))
pc
```
